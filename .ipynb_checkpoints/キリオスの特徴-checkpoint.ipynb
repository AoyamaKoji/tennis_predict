# ハードコートを対象にする

```python
Hard_match_before_2018 = match_result_before_2018_df[match_result_before_2018_df.Surface == "Hard"]
print(len(Hard_match_before_2018))
Hard_match_before_2018.head()
```

```python
Hard_match_2019 = match_result_2019_df[match_result_2019_df.Surface == "Hard"]
print(len(Hard_match_2019))
Hard_match_2019.head()
```

```python
Hard_Tournament = Tournament_match_df[Tournament_match_df.Surface == "Hard"]
print(len(Hard_Tournament))
Hard_Tournament.head()
```

```python
Hard_Career_tour = Career_tour_df[Career_tour_df.Split == "Hard"]
print(len(Hard_Career_tour))
Hard_Career_tour.head()
```

# データを削除

```python
match_before_2018_des = Hard_match_before_2018.drop(columns = ['Date','Surface','Rd', 'Rk', 'vRk', 'Player', 'Score'], axis=1)
print(len(match_before_2018_des))
match_before_2018_des.head()
```

```python
match_2019_des = Hard_match_2019.drop(columns = ['Date','Surface','Rd', 'Rk', 'vRk', 'Player', 'Score'], axis=1)
print(len(match_2019_des))
match_2019_des.head(30)
```

```python
Tour_seasons = Tour_seasons_df.drop(columns = ['Best'], axis=1)
pd.set_option('display.max_columns', 50)
print(len(Tour_seasons))
Tour_seasons.head(10)
```

```python
Challenger_seasons = Challenger_seasons_df
print(len(Challenger_seasons))
Challenger_seasons.head()
```

```python
Tournament_match = Hard_Tournament.drop(columns = ["Surface", "First", "Last", "Best", "MS"], axis=1)
print(len(Tournament_match))
Tournament_match.head(10)
```

```python
Career_tour = Hard_Career_tour.drop(columns = ["Split"], axis=1)
print(len(Career_tour))
Career_tour.head()
```
